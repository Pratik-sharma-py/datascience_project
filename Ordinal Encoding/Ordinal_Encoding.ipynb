```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/customer.csv')
```

```python
data.sample(5)
```

```python
data = data.iloc[:,2:]
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,0:2],data.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train_enc = oe.transform(X_train)
X_test_enc = oe.transform(X_test)
```

```python
X_train_enc = pd.DataFrame(X_train_enc, columns= X_train.columns)
X_test_enc = pd.DataFrame(X_test_enc, columns= X_test.columns)
X_train_enc.head(5)
X_test_enc.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
y_train_enc = le.transform(y_train)
y_test_enc = le.transform(y_test)
```

```python
y_train_enc
```

```python
y_train_enc = pd.DataFrame(y_train_enc, columns=[y_train.name])
y_test_enc = pd.DataFrame(y_test_enc, columns=[y_test.name])
y_train_enc.head(5)
y_test_enc.head(5)
```
